Fun with microCT data

Take some model data (microCT of a mouse) from https://figshare.com/articles/dataset/Mouse_micro-CT_raw_data_for_segmentation/28303637.

Open the (proprietary format) file.  (Used nibabel, https://nipy.org/nibabel/index.html)

Extract and plot slices from the data.

Use k-means clustering (a widely used unsupervised machine learning algorithm for segmenting micro-CT data, particularly in medical and materials research) to segment a slice.  

Next step would be to segment volumes (need a better computer).


```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

import os
import nibabel as nib
import matplotlib.pyplot as plt
from nibabel.testing import data_path
```

```python
#locate data file
cd Raw\ Data\ for\ microCT\ segmentation/Training/C57_#6/
```

```python
#load data
img2 = nib.load('C57-#6-MicroCT-020421 - Cardiac CT_110734-systole.nii')
```

```python
image2_data = img2.get_fdata()
```

```python
#test slice from middle of the stack
mid_slice_x = image2_data[256, :, :]
# is it the size we expected?  Yes.
print(mid_slice_x.shape)
```

```python
#Make some plots in all three directions.
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

ax[0].imshow(image2_data[255, :, :].T, origin='lower', cmap='gray')
ax[0].set_xlabel('Second dim voxel coords.', fontsize=12)
ax[0].set_ylabel('Third dim voxel coords', fontsize=12)
ax[0].set_title('First dimension, slice no. 255', fontsize=15)

ax[1].imshow(image2_data[:, 255, :].T, origin='lower', cmap='gray')
ax[1].set_xlabel('First dim voxel coords.', fontsize=12)
ax[1].set_ylabel('Third dim voxel coords', fontsize=12)
ax[1].set_title('Second dimension, slice no. 255', fontsize=15)

ax[2].imshow(image2_data[:, :, 255].T, origin='lower', cmap='gray')
ax[2].set_xlabel('First dim voxel coords.', fontsize=12)
ax[2].set_ylabel('Second dim voxel coords', fontsize=12)
ax[2].set_title('Third dimension, slice no. 255', fontsize=15)

fig.tight_layout()
```

```python
plt.hist(img_flat, bins=256)
plt.title('Basic Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn import cluster
from skimage import data

def km_clust(array, n_clusters):
    
    # Create a line array, the lazy way
    X = array.reshape((-1, 1))
    # Define the k-means clustering problem
    k_m = cluster.KMeans(n_clusters=n_clusters, n_init=4)
    # Solve the k-means clustering problem
    k_m.fit(X)

    # Get the coordinates of the clusters centres as a 1D array
    values = k_m.cluster_centers_.squeeze()

    # Get the label of each point
    labels = k_m.labels_

    return(values, labels)

# Group similar grey levels using 2 clusters
values, labels = km_clust(mid_slice_x, n_clusters = 2)

# Create the segmented array from labels and values
img_segm = np.choose(labels, values)
# Reshape the array as the original image
img_segm.shape = mid_slice_x.shape

# Group similar grey levels using 8 clusters
values, labels = km_clust(mid_slice_x, n_clusters = 8)

# Create the segmented array from labels and values
img_segm8 = np.choose(labels, values)
# Reshape the array as the original image
img_segm8.shape = mid_slice_x.shape

# Get the values of min and max intensity in the original image
vmin = mid_slice_x.min()
vmax = mid_slice_x.max()
vmin2 = img_segm.min()
vmax2 = img_segm.max()
vmin3 = img_segm8.min()
vmax3 = img_segm8.max()

fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

# Plot the original image
ax[0].imshow(mid_slice_x,cmap=plt.cm.gray, vmin=vmin, vmax=vmax)
ax[0].set_title('Original image')

# Plot the simplified color image
ax[1].imshow(img_segm, cmap=plt.cm.plasma, vmin=vmin2, vmax=vmax2)
ax[1].set_title('Poor segmentation')

# Plot the simplified color image
ax[2].imshow(img_segm8, cmap=plt.cm.plasma, vmin=vmin3, vmax=vmax3)
ax[2].set_title('Better segmentation')

fig.tight_layout()
```
